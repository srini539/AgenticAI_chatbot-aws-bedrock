```
# String slicing

x = "Hello LTM, I am a new joiner"

y = x[:12]
print(y)

z = x[-10:-2]
print(z)

print(x.upper())

print(x.lower())
print(x.strip())

a = x.replace('H',"Hi H")
print(a)

b = x.split(" ")
print(b) # returns a list
```

```
# String concatenation

s1  = "this is"
s2 = "a laptop"
s3 = "manufactured by Dell."

s4 = s1 + " " + s2 + ' ' + s3

print(s4)
```
